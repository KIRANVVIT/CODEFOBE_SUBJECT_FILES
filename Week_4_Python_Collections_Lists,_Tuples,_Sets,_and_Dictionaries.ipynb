## **1.   Introduction to Nested Lists**

A nested list is simply a list within another list. In Python, a list can hold a variety of data types, including other lists. This ability to "nest" lists enables us to create more complex structures, similar to tables or matrices, which can be useful for organizing and managing multi-level data.

**Why Use Nested Lists?**

Nested lists are useful because they allow us to group related data together in an organized way, especially when dealing with hierarchical or multi-dimensional data. Without nested lists, handling structured information in a clean and manageable way becomes challenging.

**Nested lists are particularly handy when:**

You want to represent data in a structured way (like tables, grids, or groups).
You have data that naturally forms a hierarchy (e.g., categories and subcategories, groups and subgroups).

**Real-World Example:** Consider organizing a menu for a restaurant. The menu could have different categories, such as appetizers, main courses, and desserts. Each category has a list of items.





   




```python
# Nested list for menu categories
menu = [
    ["Bruschetta", "Spring Rolls"],           # Appetizers
    ["Steak", "Pasta", "Salmon"],             # Main Courses
    ["Ice Cream", "Cheesecake", "Brownie"]    # Desserts
]

# Defining category names for easy reference
categories = ["Appetizers", "Main Courses", "Desserts"]

# Printing each category and its items
for category, items in zip(categories, menu):
    print(f"{category}: {items}")
```

**In this example:**
* The menu list contains three lists, each representing a category of the menu.
* Each sublist contains items under that specific category (e.g., appetizers, main courses, desserts).

Using this nested list structure allows us to easily access, modify, or add menu items without having a separate variable for each category.

## **1. Introduction to Tuples**
A tuple in Python is a collection data type similar to a list but with one major difference: tuples are immutable. This means that once a tuple is created, its elements cannot be modified, added, or removed. Tuples are defined by placing items inside parentheses () separated by commas.

**How to Create Tuples**

Creating a tuple is simple:

* Place elements inside parentheses, separated by commas.
* If creating a tuple with only one element, include a trailing comma to distinguish it from a regular value in parentheses.

**Scenario: Storing Dates and Times for Important Events**

Imagine you’re building an app that helps users keep track of important life events, like birthdays, anniversaries, or scheduled appointments. The purpose of the app is to store these events with specific dates and times and display them when needed. Since these events occur on fixed dates and times, it’s essential that the stored data remains unchangeable to prevent accidental modifications. This is where tuples come in handy.

**Why Use Tuples for Storing Dates and Times?**

1. Immutability: Tuples are immutable, meaning once you set a date or time in a tuple, it cannot be altered. This is ideal for dates and times since these values are typically fixed and don’t need to be updated once recorded.
2. Data Integrity: By using tuples, you ensure that each stored date (like a birthdate or anniversary) remains consistent, reducing the risk of unintended changes in your program.
3. Clarity and Simplicity: Tuples are simple structures that allow you to group related values (like year, month, day) together in one place, making it easy to understand and access date and time information.

**Suppose you have a user, John, who wants to record his birthday and a meeting appointment in the app. Here’s how the app would handle these events:**

**Storing a Fixed Birthdate:**

1. John’s birthday is on July 20, 1995. Since John’s birthdate won’t ever change, it makes sense to store it as a tuple containing the year, month, and day: (1995, 7, 20).
2. Using a tuple ensures that John’s birthday remains consistent across all instances in the app, even if other data is modified.
3. This tuple structure makes it clear what each value represents (first value is the year, second is the month, and third is the day).

**Storing a Fixed Appointment Time:**

1. John also has a project meeting scheduled at 2:30 PM (14:30 in 24-hour format) on a specific day.
2. To store the time accurately, the app uses a tuple: (14, 30, 0), where the first value represents the hour, the second the minute, and the third the seconds (set to 0 in this case because there’s no specific seconds value).
3. Like the birthdate, this tuple structure ensures that John’s meeting time remains constant in the app.

```python
# Storing John’s birthdate as a tuple (year, month, day)
birthdate = (1995, 7, 20)

# Storing John's project meeting time as a tuple (hour, minute, second)
meeting_time = (14, 30, 0)

# Printing each event with a clear description
print("John’s Important Events:")
print(f"Birthday: Year {birthdate[0]}, Month {birthdate[1]}, Day {birthdate[2]}")
print(f"Project Meeting Time: {meeting_time[0]}:{meeting_time[1]:02d}:{meeting_time[2]:02d}")
```

02d ensures that this value is always displayed as 2 digits. If it's a single digit (e.g., 5), it will be padded with a 0 in front (e.g., 05).

## **2. Accessing Elements in Tuples**

Accessing elements in a tuple is straightforward and similar to accessing elements in a list. Since tuples are indexed data structures, you can access their elements using an index, where the first element has an index of 0, the second element has an index of 1, and so on.

**Syntax for Accessing Tuple Elements**

* To access an element in a tuple, use the syntax:

   **tuple_name[index]**

* tuple_name is the name of the tuple.

* index is the position of the element you want to access (starting from 0).

## **1. Introduction to Sets and Creating Sets**

**What are Sets?**

A set is a collection of unique elements in Python, meaning it cannot contain duplicates. Sets are unordered, so the items do not have a defined order and cannot be accessed by an index like in lists or tuples. Sets are created using curly braces {}

**Why Use Sets?**

Sets are useful when you need to store unique items and want to avoid duplicates, such as unique user IDs, tags, or types of items in a list.

**Real-World Example:**

Imagine you run an online store and want to keep track of unique categories customers are interested in. Each time a customer browses a category, you add it to a set. This way, even if they browse the same category multiple times, it will only appear once in the set.



```python
# Creating a set of unique product categories viewed by customers
viewed_categories = {"electronics", "books", "furniture", "electronics"}  # Duplicate "electronics" is ignored
print("Unique categories viewed:", viewed_categories)
```

## **Scenario: Unique Login IP Addresses (Cybersecurity)**

* Imagine you are building a login security system for a website.
* Every time a user logs in, their IP address is recorded.

👉 But:A user may log in multiple times from the same IP
You only want to track unique IP addresses to detect suspicious activity


```python
login_ips = {"192.168.1.1", "192.168.1.2", "192.168.1.1", "192.168.1.3"}

print("Unique Login IPs:", login_ips)
```

## **2. Accessing Elements in Sets**

**What Does Accessing Mean?**

Since sets are unordered, you cannot access elements by an index. Instead, you can loop through the elements or check for the presence of an item using the in keyword.

**Why Use This?**

This is helpful when you don’t need the order of elements but want to quickly check if an item exists.

**Real-World Example:**
Suppose you want to check if a customer viewed a specific product category in the online store.

```python
# Set of categories viewed
viewed_categories = {"electronics", "books", "furniture"}

# Checking if a category was viewed
if "books" in viewed_categories:
    print("Customer viewed the books category.")
else:
    print("Customer has not viewed the books category.")
```

## **3. Adding and Removing Elements in Sets**

**Adding Elements**

You can add items to a set using the add() method, and any duplicate items will be ignored.

**Removing Elements**

To remove items, you can use the remove() or discard() method. If the element does not exist in the set, remove() will raise an error, while discard() will not.

**Why Use This?**

This is useful when you want to dynamically keep track of a set of unique items and need to update the collection as elements are added or removed.

**Real-World Example:**

Imagine you are keeping track of products currently in a customer’s shopping cart. You add and remove items as they add or delete products.

```python
# Products in a shopping cart
cart = {"apple", "banana", "milk"}

# Adding a new product
cart.add("bread")
print("Cart after adding bread:", cart)

# Removing an item from the cart
cart.remove("banana")
print("Cart after removing banana:", cart)
```

## **4. Set Operations (Union, Intersection, Difference)**

**Set Operations**

* Union (| or union()): Combines elements from both sets, removing any duplicates. You can use either the | symbol or the union() function to find the union of two sets.
* Intersection (& or intersection()): Keeps only elements that are present in both sets. You can use either the & symbol or the intersection() function to find the intersection.
* Difference (- or difference()): Keeps elements that are only in the first set and removes any that are in the second set. You can use either the - symbol or the difference() function to find the difference.


**Why Use These Operations?**

These operations are helpful for comparing groups to find common members, unique members, or exclusive members in each group.

**Real-World Example:** Tracking People Who Said They’d Attend vs. People Who Actually Showed Up

Imagine you’re organizing an event and keeping track of two groups of people:

* People who said they would attend (they told you in advance that they’d come).
* People who actually showed up (they came to the event).

With set operations, you can determine:

* Who said they’d attend but didn’t actually come.
* Who came to the event without telling you in advance.
* Who both said they’d attend and actually showed up.


```python
# Set of people who said they would attend
said_they_would_come = {"Alice", "Bob", "Charlie"}

# Set of people who actually showed up at the event
actually_came = {"Bob", "David"}

# 1. Union: Find all unique participants (either said they’d come or actually showed up)
print("All unique participants:", said_they_would_come | actually_came)

# 2. Intersection: Find people who both said they’d come and actually showed up
print("People who said they’d come and actually showed up:", said_they_would_come & actually_came)

# 3. Difference: Find people who said they’d come but didn’t show up
print("People who said they’d come but didn’t show up:", said_they_would_come - actually_came)
```

**Explanation of Each Set Operation**

1. Union (All Unique Participants):

* We use either the | symbol (said_they_would_come | actually_came) or the union() function **(said_they_would_come.union(actually_came))**.
* This combines everyone who either said they would come or actually came, without duplicates.
* In this example, the result will include "Alice", "Bob", "Charlie", and "David".
2. Intersection (People Who Both Said They’d Come and Actually Showed Up):

* We use either the & symbol (said_they_would_come & actually_came) or the intersection() function **(said_they_would_come.intersection(actually_came))**.
* This gives us people who are in both sets—those who said they’d come and actually showed up.
* Here, only "Bob" both said he would come and actually attended.
3. Difference (People Who Said They’d Come but Didn’t Show Up):

* We use either the - symbol (said_they_would_come - actually_came) or the difference() function **(said_they_would_come.difference(actually_came))**.
* This gives us people who said they’d come but didn’t actually attend.
* In this case, "Alice" and "Charlie" are in the result, as they said they’d attend but didn’t show up.

## **1. Introduction to Dictionaries and Creating Dictionaries**

**What are Dictionaries?**

A dictionary in Python is a collection of key-value pairs, where each key is associated with a specific value. Dictionaries use curly braces {} and provide an easy way to store and retrieve data by using unique keys.

**Why Use Dictionaries?**

Dictionaries are useful when you need to quickly look up information by a unique identifier (like a name or ID) and store related data.

**What are Dictionary Methods?**

Dictionaries in Python have built-in methods like keys(), values(), and items() that allow you to work with different parts of the data in the dictionary:

* keys(): Returns all the keys in the dictionary.
* values(): Returns all the values in the dictionary.
* items(): Returns all key-value pairs as tuples.

**Why Use These Methods?**

These methods allow you to efficiently access specific parts of a dictionary. For example, if you only need the keys, values, or both key-value pairs, these methods provide a quick and organized way to retrieve them.

**Real-World Example:**

Imagine you’re managing an employee database where each employee has a unique ID. You can store each employee's information as key-value pairs, where the key is the employee ID and the value is the employee’s details.


```python
# Creating a dictionary to store employee data
employees = {
    101: "Alice Johnson",
    102: "Bob Smith",
    103: "Charlie Davis",
    104: "Diana Prince"
}
print("Employee Database:", employees)
```

**1. keys() Method**

The keys() method returns all the keys in the dictionary. This is useful if you want to see all unique identifiers (like employee IDs) without the associated values.

```python
# Getting all employee IDs
employee_ids = employees.keys()
print("Employee IDs:", list(employee_ids))
```

**2. values() Method**

The values() method returns all the values in the dictionary. This is helpful when you need to access just the employee names without the IDs.

```python
# Getting all employee names
employee_names = employees.values()
print("Employee Names:", list(employee_names))
```

**3. items() Method**

The items() method returns all key-value pairs as tuples. This is useful when you want to work with both the keys and values together, such as when displaying a directory with IDs and names.

```python
# Getting all employee ID and name pairs as a list of tuples
employee_items = employees.items()
print("Employee Directory:", employee_items)
```

Suppose you want to find the name of an employee with a specific ID in the employee database.

```python
# Accessing the name of the employee with ID 102
employee_name = employees[102]
print("Employee with ID 102:", employee_name)
```

## **3. Adding and Modifying Key-Value Pairs**

**What Does Adding and Modifying Mean?**

You can add new key-value pairs to a dictionary, or update an existing key with a new value.

**Why Use This?**

This is useful when your data changes over time, and you need to keep your dictionary up-to-date.

**Real-World Example:**

If a new employee joins the company, or an employee’s name changes (like a nickname), you can add or update their record in the dictionary.

```python
# Adding a new employee
employees[105] = "Robert Downey Jr"
print("After adding new employee:", employees)

# Modifying an existing employee's name
employees[102] = "Robert Smith"
print("After modifying employee name:", employees)
```

## **4. Removing Key-Value Pairs**

**What Does Removing Mean?**

You can remove a key-value pair from a dictionary using the del statement or the pop() method.

**Why Use This?**

Removing items is useful when data becomes outdated or irrelevant, allowing you to keep the dictionary accurate.

**Real-World Example:**

If an employee leaves the company, you can remove their entry from the employee dictionary.

```python
# Removing an employee who left the company
del employees[103]
print("After removing employee with ID 103:", employees)
```
